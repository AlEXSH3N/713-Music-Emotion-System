```python
# Remove stop words and unnecessary words
stop_words = ['the', 'a', 'an', 'and', 'or', 'but', 'if', 'then', 'else', 'for', 'while', 'with', 'in', 'on', 'at', 'to', 'from', 'by', 'of', 'as', 'is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'do', 'does', 'did', 'will', 'would', 'should', 'can', 'could', 'may', 'might', 'must', 'shall', 'this', 'that', 'these', 'those', 'it', 'its', 'they', 'them', 'their', 'you', 'your', 'we', 'our', 'us', 'he', 'him', 'his', 'she', 'her', 'hers', 'i', 'me', 'my', 'mine', 'so', 'yes', 'yeah', 'ok', 'okay', 'oh', 'ah', 'uh', 'um', 'er', 'huh', 'eh', 'well', 'like', 'really', 'just', 'even', 'though', 'although', 'because', 'since', 'until', 'while', 'when', 'where', 'why', 'how', 'what', 'which', 'who', 'whom', 'whose', 'am', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'do', 'does', 'did', 'will', 'would', 'should', 'can', 'could', 'may', 'might', 'must', 'shall', 'go', 'goes', 'went', 'gone', 'get', 'got', 'come', 'came', 'see', 'saw', 'look', 'looked', 'say', 'said', 'tell', 'told', 'ask', 'asked', 'make', 'made', 'take', 'took', 'give', 'gave', 'put', 'set', 'let', 'keep', 'kept', 'show', 'showed', 'call', 'called', 'try', 'tried', 'find', 'found', 'want', 'wanted', 'need', 'needed', 'use', 'used', 'some', 'any', 'many', 'much', 'few', 'little', 'more', 'most', 'other', 'another', 'such', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine', 'ten', 'first', 'second', 'third', 'fourth', 'fifth', 'sixth', 'seventh', 'eighth', 'ninth', 'tenth', 'last', 'before', 'after', 'above', 'below', 'over', 'under', 'across', 'through', 'around', 'about', 'into', 'out', 'off', 'up', 'down', 'left', 'right', 'back', 'front', 'side', 'top', 'bottom', 'near', 'far', 'here', 'there', 'now', 'then', 'today', 'tomorrow', 'yesterday', 'day', 'night', 'morning', 'evening', 'afternoon', 'week', 'month', 'year', 'time', 'hour', 'minute', 'second', 'moment', 'date', 'way', 'thing', 'stuff', 'kind', 'sort', 'type', 'part', 'piece', 'bit', 'lot', 'whole', 'half', 'rest', 'all', 'every', 'each', 'both', 'either', 'neither', 'only', 'same', 'different', 'other', 'another', 'else', 'something', 'nothing', 'anything', 'everything', 'always', 'never', 'sometimes', 'often', 'usually', 'rarely', 'seldom', 'hardly', 'ever', 'again', 'once', 'twice', 'thrice', 'times', 'also', 'too', 'well', 'very', 'quite', 'rather', 'fairly', 'pretty', 'enough', 'indeed', 'surely', 'certainly', 'probably', 'maybe', 'perhaps', 'possibly', 'likely', 'unlikely', 'definitely', 'absolutely', 'completely', 'totally', 'almost', 'nearly', 'hardly', 'barely', 'scarcely', 'yet', 'still', 'already', 'soon', 'later', 'somehow', 'somewhere', 'anywhere', 'everywhere', 'anywhere', 'nowhere', 'someone', 'somebody', 'anyone', 'anybody', 'nobody', 'no', 'one', 'none', 'nothing', 'all', 'everything', 'everybody', 'everyone', 'anybody', 'anyone', 'somebody', 'someone', 'anybody', 'anything', 'something', 'nothing', 'everything', 'whatever', 'whichever', 'whoever', 'whomever', 'wherever', 'whenever', 'however', 'whatever', 'whatever', 'whichever', 'whoever', 'whomever', 'wherever', 'whenever', 'however', 'why', 'what', 'which', 'who', 'whom', 'whose', 'where', 'when', 'how', 'why', 'because', 'why', 'what', 'which', 'who', 'whom', 'whose', 'where', 'when', 'how', 'why', 'because']
data['lyrics'] = data['lyrics'].apply(lambda x: ' '.join([word for word in str(x).split() if word.lower() not in stop_words]))

# Remove unnecessary characters
data['lyrics'] = data['lyrics'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)

# Get a balance dataset for valence, energy and danceability labels
valence_bins = [0, 0.3, 0.7, 1.0]
energy_bins = [0, 0.3, 0.7, 1.0]
danceability_bins = [0, 0.3, 0.7, 1.0]

# Convert continuous values to discrete labels using the cut function
data['valence_bin'] = pd.cut(data['valence'], bins=valence_bins, labels=[0, 1, 2], right=False)
data['energy_bin'] = pd.cut(data['energy'], bins=energy_bins, labels=[0, 1, 2], right=False)
data['danceability_bin'] = pd.cut(data['danceability'], bins=danceability_bins, labels=[0, 1, 2], right=False)

balance_data = pd.DataFrame()  # Create an empty DataFrame
for valence_bin in data['valence_bin'].unique():
    for energy_bin in data['energy_bin'].unique(): 
        for danceability_bin in data['danceability_bin'].unique():
            subset = data[(data['valence_bin'] == valence_bin) & 
                          (data['energy_bin'] == energy_bin) & 
                          (data['danceability_bin'] == danceability_bin)]
            sample_subset = subset.sample(min(subset.shape[0], 50), replace=True)
            balance_data = pd.concat([balance_data, sample_subset])

# reset index
balance_data.reset_index(drop=True, inplace=True)




```
